## ConvertedCompYearly

````python
# What is your current total compensation (salary, bonuses, and perks, before taxes and deductions)
bins = public['ConvertedCompYearly'].quantile([0, .25, .5, .75, .9, .95, .99, 1]).values.tolist()
data = pd.cut(public['ConvertedCompYearly'], bins)
pd.DataFrame(data, columns=['ConvertedCompYearly']).stb.freq(['ConvertedCompYearly'])
````

````python
fig, ax = plt.subplots(figsize=(8,5))
plot_df = public[['ConvertedCompYearly']]
plot_df.boxplot(ax=ax);
````

<br>```We notice a huge amount of right skewness due to extremely high values, probably wrong entry.```

````python
desc = public.describe().loc[['25%', '75%'],['ConvertedCompYearly']]
q1 = desc.values[0][0]
q3 = desc.values[1][0]
iqr = q3-q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(f'iqr: {iqr}\nupper_bound: {upper_bound}\nlower_bound: {lower_bound}')
````

````python
plot_df = public[['ConvertedCompYearly']]
df_boxplot = plot_df[(plot_df.ConvertedCompYearly > lower_bound) & (plot_df.ConvertedCompYearly < upper_bound)]
fig, ax = plt.subplots(figsize=(8,5))
df_boxplot.boxplot(ax=ax)
plt.show()
````

<br>```Above is boxplot after removing the outliers.```

## Age

````python
# What is your age?
public.stb.freq(['Age'])
````

````python
(
    public
    [['Age','MainBranch']]
    [public.MainBranch=='I am a developer by profession']
    .stb.freq(['Age'])
)
````

<br>```48.42% of professional developers are 25-34 years old.```

````python
(
    public
    [['Age','MainBranch']]
    [public.Age=='65 years or older']
    .stb.freq(['MainBranch'])
)
````

<br>```Almost half of the respondents 65 years or older do not consider themselves professional developers.```

````python
df_plot = (
    public
    [['Age','MainBranch']]
    [public.MainBranch=='I am a developer by profession']
    .stb.freq(['Age'])
)

fig, ax = plt.subplots(figsize=(16,8))
df_plot = df_plot.iloc[:, [0, 2]].set_index('Age')
index = ['Prefer not to say',
         '65 years or older',
         '55-64 years old',
         '45-54 years old',
         '35-44 years old',
         '25-34 years old',
         '18-24 years old',
         'Under 18 years old']
df_plot = df_plot.reindex(index=index)
df_plot.plot.barh(ax = ax, legend=False, grid=True, fontsize=12)
for p in ax.patches:
    w = p.get_width()
    ax.annotate(f'{w:.2f}', (w * 1., p.get_y() + 0.1))
ax.set_ylabel(ylabel='Age', fontsize=30)
plt.show()
````
