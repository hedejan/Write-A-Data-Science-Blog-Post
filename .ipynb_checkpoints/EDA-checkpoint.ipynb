### 3. Countries with DS as most common 10 developer type

````python
tmp=(
    ds
    .groupby(['Country']).ResponseId
    .count()
    .sort_values(ascending=False)
    [:10]
)
print(f'The top 10 countries by count of DS developers are:')
[f'{i+1}. {c}' for i, c in enumerate(tmp.index.values)]
````

### 4. Education level w. stats

````python
tmp=(
    ds
    .groupby(['EdLevel']).ResponseId
    .count()
    .sort_values(ascending=False)
    .pipe(lambda x:x/x.sum()*100)
    .to_frame()
    .rename(columns={'ResponseId':'Perc'})
    .assign(CumPerc = lambda df:df.Perc.cumsum())
    .style.format(formatter={'Perc': "{:.2f}%",
                             'CumPerc':'{:.2f}%'
                          })
)
tmp
````

* 88% of DS are BSc or MSc or PhD.

### 5. Techniques for learning to code w/ stats

````python
tmp=(
    ds
    .LearnCode.str.split(';', expand=True)
    .stack()
    .value_counts()
    .pipe(lambda x:x/x.sum()*100)
    .to_frame()
    .rename(columns={0:'Perc'})
    .assign(CumPerc = lambda df:df.Perc.cumsum())
    .style.format(formatter={'Perc': "{:.2f}%",
                             'CumPerc':'{:.2f}%'
                          })
)
tmp
````

* Over 55% learned coding from school, Books / Physical media and Other online resources (ex: videos, blogs, etc).
* Least means of learning was (surprisingly for me) Coding Bootcamp.

### 6. Coding experience w/ stats

````python
tmp=(
    ds
    [['YearsCode']]
    .replace({r'More':50.5, 'Less':0.5, 'nan':np.nan}, regex=True)
    .dropna()
    .YearsCode
    .astype(float)
    .pipe(lambda x:pd.cut(x, [0,10,20,30,40,50,60]))
    .pipe(lambda x:x.groupby(x))
    .count()
    .pipe(lambda x:x/x.sum()) *100
)
tmp
````

* Over 80% of DS have 20 years or less of coding experience including any education with almost 49% with 10 years or less.

````python
tmp=(
    ds
    [['YearsCodePro']]
    .replace({r'More':50.5, 'Less':0.5, 'nan':np.nan}, regex=True)
    .dropna()
    .YearsCodePro
    .astype(float)
    .pipe(lambda x:pd.cut(x, [0,10,20,30,40,50,60]))
    .pipe(lambda x:x.groupby(x))
    .count()
    .pipe(lambda x:x/x.sum()) *100
)
tmp
````

* Over 90% of DS have 20 years or less of professional coding experience NOT including any education with over 73% with 10 years or less.

### 7. Most common Org. size, Lang., DB, Platform, webframe, Misc Tech, Tool Tech, Dev Env, OpSys

#### OrgSize

````python
tmp=(
    ds
    ['OrgSize']
    .pipe(lambda x:x.groupby(x))
    .count()
    # .value_counts()
    .pipe(lambda x:x/x.sum()*100)
    .sort_values(ascending=False)
)
tmp
````

````python
fig, ax = plt.subplots(figsize=(16,8))
tmp=(
    ds
    ['OrgSize']
    .replace({r' employees':'', ' to ':'-', 'I don’t know':np.nan, ' or more':'+',r'Just me.+':'1'}, regex=True)
    .dropna()
    .pipe(lambda x:x.groupby(x))
    .count()
    .loc[['1', '2-9', '10-19', '20-99', '100-499', '500-999', '1,000-4,999', '5,000-9,999', '10,000+', ]]
    .plot(kind='bar', ax=ax, grid=True, rot=0)
)
ax.set_xlabel('')
ax.set_ylabel('Count', fontsize=12)
fig.suptitle('Organization Size', fontsize=18);
````

* Most DS work in mid (20-99, probably startups) to large scale (10k+) organizations

#### Languages

````python
tmp = (
    ds
    ['LanguageHaveWorkedWith']
    .str.split(';', expand=True).stack().reset_index(drop=True)
    .value_counts(ascending=True)
    .pipe(lambda x:x/sum(x)*100)
    .sort_values(ascending=False)
    .to_frame()
    .rename(columns={0:'Perc'})
    .assign(CumPerc = lambda df:df.Perc.cumsum())
)
display(tmp)

fig, ax = plt.subplots(figsize=(16,12))
y =tmp['Perc'].sort_values(ascending=True)
y.plot.barh(ax=ax,)
for i, v in enumerate(y):
    ax.text(v, i-0.25, str(f'{v:.2f}%'), color='black', fontweight='light')
plt.show()
````

* Not surprisingly, Python was the top used language for coding followed by SQL and JavaScript and Bash/Shell constituting around 48%.

````python
(
    ds
    [['Sexuality']]
    .replace('Prefer to self-describe:',np.nan)
    .assign(Straight_Heterosexual=ds.Sexuality.str.contains('Straight',na=False, regex=True))
    .assign(Gay_Lesbian=ds.Sexuality.str.contains('Gay',na=False, regex=True))
    .assign(Bisexual=ds.Sexuality.str.contains('Bisexual',na=False, regex=True))
    .assign(Queer=ds.Sexuality.str.contains('Queer',na=False, regex=True))
    .assign(None_=ds.Sexuality.str.contains('Prefer',na=False, regex=True))
    .drop(columns=['None_'])
    .assign(Sum = lambda df:df.iloc[:,1:].sum(axis=1))
    .assign(Sexuality = ds.Sexuality)
    .pipe(lambda df:df[df.Sum > 1])
)
````
