### Cleaning

```python
# clean column names
clean.columns = [c.split('Have')[0].replace('NEW','') for c in df.columns]
```

```python
# MainBranch
clean.MainBranch = (
    clean
    .MainBranch
    .replace({r'.*developer.*':'developer',r'.*student.*':'student',r'.*hobby':'hobby',r'None.*':'None',
             }, regex=True)
)

# Employment
clean.Employment = (
    clean
    .Employment
    .replace({r'.*full-time':'full_time',r'.*part-time':'part_time',
              r'Not employed.*':'unemployed',r'Independent.*':'self-employed',
             }, regex=True)
)

# EdLevel
clean.EdLevel= (
    clean
    .EdLevel
    .replace({r'Master.*':'Masters',r'Professional.*':'Prof_degree',
              r'Bachelor’s.*':'Bachelors',r'Other doctoral.*':'PhD',
              r'Some college.*':'Non_degree',r'Secondary.*':'Secondary_school',
              r'.*Associate degree.*':'Associate_degree',r'.*elementary school.*':'Prim_school',
                                   },  regex=True)
)

# Age1stCode

def mean(x,y):
    return (x+y)/2

replace = {
    'Younger than 5 years':mean(0,5),
    '5 - 10 years':mean(5,10),
    '11 - 17 years':mean(11,17),
    '18 - 24 years':mean(18,24),
    '25 - 34 years':mean(25,34),
    '35 - 44 years':mean(35,44),
    '45 - 54 years':mean(45,54),
    '55 - 64 years':mean(55,64),
    'Older than 64 years':65,
}

clean.Age1stCode = (
    clean
    .Age1stCode
    .replace(replace)
)

# OrgSize
replace = {
    'Just me - I am a freelancer, sole proprietor, etc.':1,
    '2 to 9 employees':mean(2,9),
    '10 to 19 employees':mean(10,19),
    '20 to 99 employees':mean(20,99),
    '100 to 499 employees':mean(100,499),
    '500 to 999 employees':mean(500,999),
    '1,000 to 4,999 employees':mean(1000, 4999),
    '5,000 to 9,999 employees':mean(5000,9999),
    '10,000 or more employees':10_000,
    'I don’t know':np.nan
}
clean.OrgSize = (
    clean
    .OrgSize
    .replace(replace)
)

# LearnCode
replace = {
    'Online Courses or Certification':'Online Courses',
    'Other online resources \(ex: videos, blogs, etc\)':'Other online resources',
    'Other \(please specify\):':'Other',
    'Friend or family member':'Friend_family member',
    'Books / Physical media':'Books_Physical media',
}
clean.LearnCode = (
    clean
    .LearnCode
    .replace(replace, regex=True)

)

# YearsCode
replace = {
    r'More than.*':55,
    r'Less than.*':.5,
    
}

clean.YearsCode= (
    clean
    .YearsCode
    .replace(replace, regex=True)
    .astype(float)
)

# YearsCodePro
replace = {
    r'More than.*':55,
    r'Less than.*':.5,
}

clean.YearsCodePro= (
    clean
    .YearsCodePro
    .replace(replace, regex=True)
    .astype(float)
)

# Country

clean.Country = (
    clean
    .Country
    .replace('United States of America', 'USA')
    .replace('United Kingdom of Great Britain and Northern Ireland','UKI')
)

# Currency
clean.Currency=(
    clean
    .Currency
    .str.split('\t').str[0]
    .str.split(' ').str[0]
)

# OpSys
clean.OpSys = (
    clean
    .OpSys
    .replace('Other (please specify):','Other')
    .replace('Windows Subsystem for Linux (WSL)', 'WSL')
    .replace('Linux-based', 'Linux')
)

# Stuck
clean.Stuck=(
    clean
    .Stuck
    .replace({' \(please name\):':''}, regex=True)
    .replace({' \(please specify\):':''}, regex=True)
)

# SOSites
replace = {
    r'Stack Overflow for Teams.*':'SO for Teams',
    'I have never.*':'Never visited SO'
          }

clean.SOSites=(
    clean
    .SOSites
    .replace(replace, regex=True)
)

# SOVisitFreq
replace = {
    'Less than once per month or monthly':1,
    'A few times per month or weekly':2,
    'A few times per week':3,
    'Daily or almost daily':4,
    'Multiple times per day':5,
}

clean.SOVisitFreq=(
    clean
    .SOVisitFreq
    .replace(replace)
)

# SOPartFreq
replace = {
    'I have never participated in Q&A on Stack Overflow':0,
    'Less than once per month or monthly':1,
    'A few times per month or weekly':2,
    'A few times per week':3,
    'Daily or almost daily':4,
    'Multiple times per day':5,
    }

clean.SOPartFreq = (
    clean
    .SOPartFreq
    .replace(replace)
)

# SOComm
replace = {
    'No, not at all':0,
    'No, not really':1,
    'Neutral':2,
    'Yes, somewhat':3,
    'Yes, definitely':4,
    'Not sure':np.nan,
    }

clean.SOComm = (
    clean
    .SOComm
    .replace(replace)
)

# SOAccount
clean.SOAccount=(
    clean
    .SOAccount
    .replace("Not sure/can't remember",np.nan)
    .replace({'Yes': 1, 'No':0})
)

# OtherComms
clean.OtherComms =(
    clean
    .OtherComms
    .replace({'Yes': 1, 'No':0})
)

# Age
replace = {
    'Under 18 years old':mean(0,18),
    '18-24 years old':mean(18,24),
    '25-34 years old':mean(25,34),
    '35-44 years old':mean(35,44),
    '45-54 years old':mean(45,54),
    '55-64 years old':mean(55,64),
    '65 years or older':65,
    'Prefer not to say':np.nan,
}

clean.Age= (
    clean
    .Age
    .replace(replace, )
)
```

### Outliers

```python
# outlier func

def outliers(data):
    mean, std = np.mean(data), np.std(data)
    cut_off = std * 3
    lower, upper = mean -cut_off, mean +cut_off
    return lower, upper
```

```python
# remove extreme (outlier) values from num columns
lower, upper = outliers(clean['ConvertedCompYearly'])
print(clean.shape)
clean = clean[(clean['ConvertedCompYearly']>lower) & (clean['ConvertedCompYearly']<upper)]
print(clean.shape)
```

### Missing Values

```python
# remove columns with NaN's above threshold, outliers
data = clean.isna().sum()
lower, upper = outliers(data.values)
print(data.shape)
data = data[(data>lower)&(data<upper)]
print(data.shape)
```

```python
i = clean.columns
clean = clean[data.index]
ii = clean.columns
diff = np.setdiff1d(i,ii)
print(f'The following columns were dropped:\n{diff}')
```

```python
# remove rows with NaN's above threshold, outliers
data = clean.isna().sum(axis=1)
lower, upper = outliers(data.values)
print(data.shape)
data = data[(data>lower)&(data<upper)]
print(data.shape)
```

```python
i = len(clean)
clean = clean.filter(items = data.index, axis=0)
ii = len(clean)
diff = np.setdiff1d(i,ii)
print(f'{diff} rows were dropped.')
```

```python
# re-inspect NaN's visually
msno.matrix(clean);
```

#### Imputer missing features

```python
from sklearn.impute import SimpleImputer
```

```python
def encode_impute(df):
    
    num_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    cat_imputer = SimpleImputer(missing_values=np.nan, strategy='median')

    obj_delimiter = []
    obj_no_delimiter = []
    for col in df.select_dtypes('O').columns:
        if df[col].str.contains(';',na=False).any():
            obj_delimiter.append(col)
        else:
            obj_no_delimiter.append(col)
            
    df_num = df.select_dtypes('number')
    df_num_imputed = pd.DataFrame(num_imputer.fit_transform(df_num), columns=df_num.columns)
    
    df_cat_sep = pd.concat([df[col].str.get_dummies(sep=';') for col in obj_delimiter], axis=1)
    df_cat_sep_imputed = pd.DataFrame(num_imputer.fit_transform(df_cat_sep), columns=df_cat_sep.columns)
    
    df_cat_no_sep = pd.get_dummies(df[obj_no_delimiter])
    df_cat_no_sep_imputed = pd.DataFrame(num_imputer.fit_transform(df_cat_no_sep), columns=df_cat_no_sep.columns)
    
    encoded_imputed = pd.concat([df_cat_sep_imputed,df_cat_no_sep_imputed, df_num_imputed], axis=1)
    
    return encoded_imputed

encoded_imputed = encode_impute(clean)
```

## Modelling <a class="anchor" id="Modelling"></a>

### Feature Selection

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
X = encoded_imputed.drop(columns='ConvertedCompYearly')
y = encoded_imputed['ConvertedCompYearly'].values
```

```python
importances = mutual_info_regression(X,y)
```

```python
feat_importances = pd.Series(importances, X.columns).sort_values(ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(15,200))
feat_importances.plot(kind='barh', color='teal', ax=ax, grid=True)
max_=.5
step=max_/20
ax.set_xticks(ticks = np.arange(0, max_+step, step), minor=False)
plt.show()
```

```python
important_feat=(
    pd.DataFrame({'features':feat_importances.index,'importances':feat_importances.values})
    .sort_values(by='importances',ascending=False)
    .reset_index(drop=True)
    .pipe(lambda x:x[x.importances > .025])
    .features.tolist()
    
)
len(important_feat)
```

```python
X_imp = X[important_feat]
```

### Split featuresset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imp, y, shuffle=True, test_size=0.33)
```

### Ffeatures Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Selecting the estimator

```python
from sklearn import linear_model
```

```python
reg = linear_model.Lasso(alpha=0.1)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

## Evaluation <a class="anchor" id="Evaluation"></a>

```python
from sklearn.metrics import explained_variance_score, mean_absolute_error, r2_score
```

```python
explained_variance_score(y_test, y_pred), mean_absolute_error(y_test, y_pred),  r2_score(y_test, y_pred)
```
